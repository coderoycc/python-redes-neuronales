```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

```

## CREAMOS EL DATASET

```python
# cantidad de datos
n = 500
#numero de caracteristicas
p = 2

X, Y = make_circles(n_samples=n, factor=0.5, noise=0.05) 
    #factor es la distancia entre los circulos, noise es el ruido (variabilidad)

# Pone condicion para cuando Y sea 0 (caracteristica 1)
plt.scatter(X[Y == 0, 0], X[Y == 0, 1], c='blue')

# Cuando Y sea 1 (caracteristica 2)
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], c='red')
plt.show()
```

```python
# Clase de la capa de la red
class neural_layer():
  '''
  @param n_con: numero de conexiones que entran a nuestra capa
  @param n_neu: numero de neuronas que existen en esta capa
  @param act_f: funcion de activacion de la neurona 
  ''' 
  def __init__(self, n_con, n_neu, act_f):
    self.act_f = act_f
    self.b = np.random.rand(1, n_neu) * 2 - 1
      #Vector columna con la cantidad de neuronas
    
    self.W = np.random.rand(n_con, n_neu) * 2 - 1
      #matriz de conexiones X neuronas
```

```python
# Funciones de activacion (sigmoide, derivada(sigmoide))
sigm = (lambda x: 1/(1 + np.e**(-x)),
        lambda x: x*(1-x))

# Funcion RELU
relu = lambda x: np.maximum(0,x)

_x = np.linspace(-5, 5, 100) #creamos 100 valores de -5 a 5
plt.plot(_x, sigm[1](_x))
```

```python
layer0 = neural_layer(p, 4, sigm)
layer1 = neural_layer(4, 8, sigm)

def create_nn(topologia, act_f):
  nn = [] #todas las capas
  for i, layer in enumerate(topologia[:-1]): #Todas menos la Ãºlitma
    nn.append(neural_layer(topologia[i], topologia[i+1], act_f))

  return nn
```

```python
topologia = [2, 4, 8, 16, 8, 4, 1] #N conexiones de la red
red_neuronal = create_nn(topologia, sigm)


#entrenamos a la red neuronal
def train(red_neuronal, X, Y, func_coste, lr=0.5): #lr: learning rate 
  out = [(None, X)]
  # Forward pass "@" es una multiplicacion matricial
  for l, layer in enumerate(red_neuronal):
    z = out[-1][1] @ red_neuronal[l].W + red_neuronal[l].b
    a = red_neuronal[l].act_f[0](z)
    out.append((z,a))

  print(out[-1][1])

func_coste = (lambda Yp, Yr: np.mean((Yp - Yr) ** 2), 
              lambda Yp, Yr: (Yp - Yr))
train(red_neuronal, X, Y, func_coste, 0.5)
```
